```python
l=[2,4,6,8,3,5,8,9]
n=len(l)//2
sum1=0
sum2=0
for i in range(0,n):
    sum1+=l[i]
for i in range(n,len(l)):
    sum2+=l[i]
print("first half sum->",sum1,"second half sum->",sum2)



```

```python
l=[[1,2,3],[4,5,6],[7,8,9]]
sum=0
max=0;
for i in range(0,len(l)):
    for j in range(0,len(l[i])):
        if max<len[i][j]:
            max=len[i][j]
            sum+=max
```
