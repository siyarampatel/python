```python
# update the list
l=[3,4,5,3,2,2]
l.append("hello")
# it will add this list in an single index at last
l.append([4,5,4,3,2])
print(l)
print(l[7][2])
```

```python
l=[3,4,5,3,2,2]
l.insert(2,35)
# even if we insert form right to left indexing then also it will shift from left to right
l.insert(-3,55)
print(l)
```

```python
l=[3,4,5,3,2,2]
# for adding multiple elements in different indexes in last
l.extend([5,4,3,2])
print(l)
```

```python
# works as extend
l=[3,4,5,3,2,2]
l=l+[5,6,5,4,4]
print(l)
```

```python
l=[3,4,5,3,2,2]
del l[4]
print(l)
l.pop(4)
print(l)
```

```python
l=[3,4,5,3,2,2]
# it can return the value which is deleted
x=l.pop(2)
print(x)
print(l)
```

```python
l=[3,4,5,3,2,2]
l.remove(3)
print(l)
```
