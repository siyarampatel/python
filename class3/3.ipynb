```python
#while loop
# count digit of an number
n=int(input("give n"))
c=0;
while(n!=0):
    n=n//10
    c=c+1
print("The no. of digits is =>",c)
```

```python
# sum of digits
n=int(input("give n"))
sum=0
t=0
while(n!=0):
    t=n%10
    sum=sum+t
    n=n//10
print("The sum is->",sum)
```

```python
n=int(input("give n"))
rev=0
t=0
while(n!=0):
    t=n%10
    rev=rev*10+t
    n=n//10
print("Reversed->",rev)
```

```python
# narcissistic number
n=int(input("give n"))
comp=n
m=n
c=0
t=0
sum=0
while(n!=0):
    n=n//10
    c=c+1
while(comp!=0):
    t=comp%10
    sum=sum+t**c
    comp=comp//10
if sum==m:
    print("Armstrong number")
else:
    print("Not")



  
```
