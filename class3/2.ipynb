```python
# prime number code
n=int(input("Give an number for checking it's an prime number or not"))
for i in range(2,n):
    if(n%i==0):
     print("Not prime")
     break
else:
     print("Prime")
```

```python
# even and odd factors count of an number 
n=int(input("Give n"))
odd=0
even=0
for i in range(1 , n+1):
    if(n%i==0 and i%2==0):
        even=even+1
    elif(n%i==0 and i%2!=0):
         odd=odd+1
print(even,"-> even")
print(odd,"-> odd")
```
