```python
# WAP which will print fizz if the number is divisible by 3 
# Buzz if no id divisibel by 5
# fizz buzz if both 3 and 5
for i in range(1,31,1):
    if(i%3==0):
        print("FIZZ","->",i)
    elif(i%5==0):
        print("BUZZ","->",i)
    elif(i%3==0 and i%5==0):
        print("FIZZBUZZ","->",i)
    else:
        print(i);
```
